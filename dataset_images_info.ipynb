```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
def get_image_info(directory, num_samples=5):
    files = os.listdir(directory)
    samples = files[:num_samples]
    for sample in samples:
        img_path = os.path.join(directory, sample)
        img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        if img is not None:
            print(f"Image: {sample}")
            print(f" - Shape: {img.shape}")
            print(f" - Data type: {img.dtype}")
        else:
            print(f"Failed to read image: {sample}")
```

```python
# Define the directories
train_images_dir = 'dataset/train/images'
train_segments_dir = 'dataset/train/segments'
train_lanes_dir = 'dataset/train/lane'

print("Training Images Info:")
get_image_info(train_images_dir)

print("\nTraining Segments Info:")
get_image_info(train_segments_dir)

print("\nTraining Lanes Info:")
get_image_info(train_lanes_dir)
```

```python
def show_image_with_masks(images_dir, segments_dir, lanes_dir, num_samples=10):
    image_files = os.listdir(images_dir)
    segment_files = os.listdir(segments_dir)
    lane_files = os.listdir(lanes_dir)
    
    samples = image_files[:num_samples]
    
    for i, sample in enumerate(samples):
        img_path = os.path.join(images_dir, sample)
        seg_path = os.path.join(segments_dir, sample)
        lane_path = os.path.join(lanes_dir, sample)
        
        image = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        segment = cv2.imread(seg_path, cv2.IMREAD_UNCHANGED)
        lane = cv2.imread(lane_path, cv2.IMREAD_UNCHANGED)
        
        if image is not None and segment is not None and lane is not None:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for displaying with matplotlib
            
            plt.figure(figsize=(12, 4))
            
            plt.subplot(1, 3, 1)
            plt.imshow(image)
            plt.title(f"Image: {sample}")
            
            plt.subplot(1, 3, 2)
            plt.imshow(segment, cmap='gray')
            plt.title(f"Segment Mask: {sample}")
            
            plt.subplot(1, 3, 3)
            plt.imshow(lane, cmap='gray')
            plt.title(f"Lane Mask: {sample}")
            
            plt.show()
        else:
            print(f"Failed to read one or more files for: {sample}")
```

```python
# Define the directories
train_images_dir = 'dataset/train/images'
train_segments_dir = 'dataset/train/segments'
train_lanes_dir = 'dataset/train/lane'
```

```python
# Show images with their masks
show_image_with_masks(train_images_dir, train_segments_dir, train_lanes_dir, num_samples=10)
```
